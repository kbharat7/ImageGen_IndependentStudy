```python
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
import os


# Initialize a BlobServiceClient
connect_str = ""  # Replace with your storage account connection string
blob_service_client = BlobServiceClient.from_connection_string(connect_str)

# Name of your container
container_name = 'imagegeneration'
folder_name = 'Large'
# The local directory to download to
local_path = "./Cleaned_Images/"
os.makedirs(local_path, exist_ok=True)

try:
    # Create a container client
    container_client = blob_service_client.get_container_client(container_name)
    
    # List blobs in container with specific prefix (folder)
    blob_list = container_client.list_blobs(name_starts_with=folder_name)
    for blob in blob_list:
        # Create a blob client for the blob
        blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob.name)

        # Create a local file path inside the local directory
        local_file_path = os.path.join(local_path, blob.name)
        # Create directories if they don't exist
        os.makedirs(os.path.dirname(local_file_path), exist_ok=True)

        # Download blob to local path
        with open(local_file_path, "wb") as download_file:
            download_blob = blob_client.download_blob()
            download_file.write(download_blob.readall())

        print(f"Downloaded blob to {local_file_path}")
except Exception as ex:
    print('Exception:')
    print(ex)
```
